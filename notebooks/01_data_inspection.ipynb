```python
import pandas as pd

# set display options for better viewing
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 1000)
```

```python
file_path = "../data/raw/day_ahead_prices_germany_2023_2026.csv"

df = pd.read_csv(file_path, sep=";")
df.head()
```

```python
df = df[[
    "Start date",
    "Germany/Luxembourg [€/MWh] Calculated resolutions"
]]

df.head()
```

```python
df.columns = ["timestamp", "price"]

df.head()
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"])

df.info()
df.head()
```

```python
df = df.sort_values(by="timestamp").reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.tail()
```

```python
full_range = pd.date_range(
    start=df["timestamp"].min(),
    end=df["timestamp"].max(),
    freq="h"
)

len(full_range)
```

```python
len(df)
```

```python
missing_timestamps = full_range.difference(df["timestamp"])

len(missing_timestamps)
```

```python
missing_timestamps[:10]
```

```python
df["timestamp"].duplicated().sum()
```

```python
df[df["timestamp"].duplicated(keep=False)]
```

```python
df = df.groupby("timestamp", as_index=False).mean()
```

```python
df["timestamp"].duplicated().sum()
```

```python
len(df)
df = df.set_index("timestamp")
df = df.sort_index()

print(df.head())
print(df.columns)
print(df.shape)
```
